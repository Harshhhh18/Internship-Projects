```python
import numpy as np
import pandas as pd

# Original DataFrame fine
# (Assuming fine is defined and contains your data)

# List of date range conditions
date_ranges = [
    
    {"startday": 1, "start_month": 9, "endday": 31, "endmonth": 10},
    {"startday": 1, "start_month": 10, "endday": 30, "endmonth": 11},
    
    
    {"startday": 1, "start_month": 9, "endday": 4, "endmonth": 10},
    {"startday": 1, "start_month": 10, "endday": 4, "endmonth": 11},
    {"startday": 1, "start_month": 11, "endday": 4, "endmonth": 12},
    
    
    {"startday": 1, "start_month": 9, "endday": 30, "endmonth": 9},
    {"startday": 1, "start_month": 10, "endday": 31, "endmonth": 10},
    {"startday": 1, "start_month": 11, "endday": 30, "endmonth": 11},
    {"startday": 1, "start_month": 9, "endday": 4, "endmonth": 12}
    
]

# Create a list to store DataFrames
result_dfs = []

for date_range in date_ranges:
    startday = date_range["startday"]
    start_month = date_range["start_month"]
    endday = date_range["endday"]
    endmonth = date_range["endmonth"]
    
    fine["rain"]=round(fine["rain"],2)

    
    filtered_df = fine[((fine["day"] >= startday) & (fine["month"] >= start_month)) | (fine["month"] >= start_month+1)] 
    #filtered_df_2= fine[((fine["day"] >= 1) & (fine["month"] >= start_month+1))]
    #filtered_df=pd.concat([filtered_df_1,filtered_df_2])
    #filtered_df=filtered_df.drop_duplicates()
    
    
    filtered_df = filtered_df[((filtered_df["day"] <= endday)&(filtered_df["month"] <= endmonth))| (filtered_df["month"] <= endmonth-1)]
    #filtered_df = filtered_df[(((filtered_df["day"] >= startday) & (filtered_df["month"] >= start_month))&(filtered_df["year"]==filtered_df["uwy"]+1))  |
    #              (((filtered_df["day"] <= endday)&(filtered_df["month"] <= endmonth))&(filtered_df["year"]==filtered_df["uwy"]+1)) ]  
    
    
    filtered_df["cap_rain"]=np.where(filtered_df['rain']>100,100,filtered_df['rain'])
    
    filtered_df['rain1_uncap'] = filtered_df.groupby(['Lat', 'Long', 'year'])["rain"].shift(-1)
    filtered_df['rain2_uncap'] = filtered_df.groupby(['Lat', 'Long', 'year'])["rain"].shift(-2)
    filtered_df['rain3_uncap'] = filtered_df.groupby(['Lat', 'Long', 'year'])["rain"].shift(-3)
    filtered_df['rain4_uncap'] = filtered_df.groupby(['Lat', 'Long', 'year'])["rain"].shift(-4)
    filtered_df['rain5_uncap'] = filtered_df.groupby(['Lat', 'Long', 'year'])["rain"].shift(-5)
    filtered_df['rain6_uncap'] = filtered_df.groupby(['Lat', 'Long', 'year'])["rain"].shift(-6)
    
    filtered_df['rain1_cap'] = filtered_df.groupby(['Lat', 'Long', 'year'])["cap_rain"].shift(-1)
    filtered_df['rain2_cap'] = filtered_df.groupby(['Lat', 'Long', 'year'])["cap_rain"].shift(-2)
    filtered_df['rain3_cap'] = filtered_df.groupby(['Lat', 'Long', 'year'])["cap_rain"].shift(-3)
    filtered_df['rain4_cap'] = filtered_df.groupby(['Lat', 'Long', 'year'])["cap_rain"].shift(-4)
    filtered_df['rain5_cap'] = filtered_df.groupby(['Lat', 'Long', 'year'])["cap_rain"].shift(-5)
    filtered_df['rain6_cap'] = filtered_df.groupby(['Lat', 'Long', 'year'])["cap_rain"].shift(-6)
    
    filtered_df['uncap_sum7dayRain']=filtered_df['rain']+filtered_df['rain1_uncap']+filtered_df['rain2_uncap']+filtered_df['rain3_uncap']+filtered_df['rain4_uncap']+filtered_df['rain5_uncap']+filtered_df['rain6_uncap']
    filtered_df['uncap_sum5dayRain']=filtered_df['rain']+filtered_df['rain1_uncap']+filtered_df['rain2_uncap']+filtered_df['rain3_uncap']+filtered_df['rain4_uncap']
    filtered_df['uncap_sum3dayRain']=filtered_df['rain']+filtered_df['rain1_uncap']+filtered_df['rain2_uncap']
    
    
    filtered_df['cap_sum7dayRain']=filtered_df['cap_rain']+filtered_df['rain1_cap']+filtered_df['rain2_cap']+filtered_df['rain3_cap']+filtered_df['rain4_cap']+filtered_df['rain5_cap']+filtered_df['rain6_cap']
    filtered_df['cap_sum5dayRain']=filtered_df['cap_rain']+filtered_df['rain1_cap']+filtered_df['rain2_cap']+filtered_df['rain3_cap']+filtered_df['rain4_cap']
    filtered_df['cap_sum3dayRain']=filtered_df['cap_rain']+filtered_df['rain1_cap']+filtered_df['rain2_cap']
    
    
    # Calculate your desired aggregations on the filtered DataFrame
    maxsum5_filtered = filtered_df.groupby(['Lat', 'Long', 'year']).agg(start_dt=('Date', 'min'),end_dt=('Date', 'max'),
                                                                duration_days=('day', 'count'),  
                                             uncap_rain_count=('rain', 'count'),
                                                  uncap_max_singleDAY=('rain', 'max'),
                                                        uncap_ERI_3day=('uncap_sum3dayRain', 'max'),            
                                                      uncap_ERI_5day=('uncap_sum5dayRain', 'max'),
                                                      uncap_ERI_7day=('uncap_sum7dayRain', 'max'),
                                                      uncap_duration_Rain_sum=('rain', 'sum'),
                                                              
                                                  cap_rain_count=('cap_rain', 'count'),
                                                  cap_max_singleDAY=('cap_rain', 'max'),
                                                            cap_ERI_3day=('cap_sum3dayRain', 'max'),      
                                                      cap_ERI_5day=('cap_sum5dayRain', 'max'),
                                                      cap_ERI_7day=('cap_sum7dayRain', 'max'),
                                                      cap_duration_Rain_sum=('cap_rain', 'sum')).reset_index()
   
    
    fine=fine1
    # Add a new column for the date range
    maxsum5_filtered['DateRange'] = f"{startday}-{start_month} to {endday}-{endmonth}"

    # Append the filtered DataFrame to the result_dfs list
    result_dfs.append(maxsum5_filtered)
# Concatenate the list of DataFrames into a single DataFrame
maxsum5_combined = pd.concat(result_dfs)

# Now you have maxsum5_combined
```
