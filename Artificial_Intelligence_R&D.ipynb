```python
df = df.drop(['Trend Adjusted Average', 'Max Loss ', 'Unnamed: 12',  'Strike 2  (>= mm)',
       'Strike 3  (>= mm)'], axis = 1)
```

```python
df
```

```python
repeat_count = 2
```

```python
for i in range(len(df)):
    if pd.isnull(df.loc[i, 'State']):
        df.loc[i, 'State'] = df.loc[i - 1, 'State'] if i % repeat_count != 0 else df.loc[i - 1 - (i % repeat_count), 'State']
```

```python
df
```

```python
df = df[df['State'] != 'PORTFOLIO']
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_encoded = pd.get_dummies(df, columns =['State', 'District'], drop_first = True)
```

```python
df.columns
```

```python
X = df_encoded.drop(columns = ['Strike 1\n (>= mm)'])
```

```python
y = df_encoded[[ 'Strike 1\n (>= mm)']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

```python
X_train.dtypes
```

```python
y_train.dtypes
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
```

```python
mse
```

```python
r2 = r2_score(y_test, y_pred)
```
